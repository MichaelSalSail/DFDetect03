```python
import face_recognition
from PIL import Image
import cv2
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def all_faces(input_image):
    locations = face_recognition.face_locations(input_image)
    print("Below are all the faces face_recognition detected.")
    for x in locations:
        top, right, bottom, left = x
        face_image = input_image[top:bottom, left:right]
        pil_image = Image.fromarray(face_image)
        pil_image.show()
```

```python
# example = cv2.imread("images/examples/adams.jpg")
blur=face_recognition.load_image_file("Examples/Blurs.png")
unique_faces=face_recognition.load_image_file("Examples/Different Faces.png")
# example_rgb = cv2.cvtColor(example, cv2.COLOR_BGR2RGB)
plt.imshow(blur[:, :, ::-1 ])
plt.show()
```

```python
all_faces(blur)
```

```python
def save_all_faces(names_df, current, destination):
    for i in range(0,40):
        # full_1=destination+input_image
        full_1=destination+'/'+str(names_df["Name"].iloc[i])
        # full_2=current+input_image
        # load_image=face_recognition.load_image_file(full_2)
        load_image=face_recognition.load_image_file(current+'/'+str(names_df["Name"].iloc[i]))
        locations = face_recognition.face_locations(load_image)
        for x in locations:
            top, right, bottom, left = x
            face_image = load_image[top:bottom, left:right]
            pil_image = Image.fromarray(face_image)
            pil_image.save(full_1)
            pil_image.show()
```
