```python
# DeepFace was created by FaceBook. The model incorporates other models like VGG-Face,
# Google Facenet, and OpenFace for high quality performance. It's a complicated model 
# to understand, which is why we went with OpenFace for transfer learn. However, the 
# strong facial analysis module provided by DeepFace may prove useful for our project.
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1_path = 'example-person.png'
img1 = cv2.imread(img1_path)
plt.imshow(img1[:, :, ::-1 ]) 
plt.show() 
```

```python
# Facial Feature Analysis
# If you're running this for the first time, it will download 1.5GB in pretrained weights.
obj = DeepFace.analyze(img_path = 'example-person.png', actions = ['age', 'gender', 'race', 'emotion'])
print("    Age:", obj["age"])
# Race -> asian, white, middle eastern, indian, latino, or black.
print("   Race:", obj["dominant_race"])
# Emotion -> angry, fear, neutral, sad, disgust, happy, and surprise.
print("Emotion:", obj["dominant_emotion"])
print(" Gender:", obj["gender"])
```
