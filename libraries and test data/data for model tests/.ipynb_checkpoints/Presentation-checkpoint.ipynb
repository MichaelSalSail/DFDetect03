This notebook will print data visualizations used for the **11.23 Mini-Presentations**. All test runs outputs were saved to excel sheets for data analysis.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set up all dataframes
eyes_df = pd.read_csv("All Labels/eyeblinks.csv", header=0)
beard_df = pd.read_csv("All Labels/beards.csv", header=0)
shades_df = pd.read_csv("All Labels/sunglasses_shades.csv", header=0)
glasses_df = pd.read_csv("All Labels/sunglasses_glasses.csv", header=0)
```

```python
eyes_df
```

```python
eyes_df[['Gender']].value_counts(sort=False).plot.pie(labels=('Women','Men'),autopct='%1.1f%%')
plt.title('Gender')
plt.show()
```

```python
eyes_df[['Race']].value_counts(sort=False).plot.pie(labels=('Non-White','White'),autopct='%1.1f%%', colors = ('#93C572', '#FFF8DC'))
plt.title('Race')
plt.show()
```

```python
# 10 * 10 sec videos = 1500 frames
# Avg. number of frames wrong per video
print('Accuracy rate of',((150-(eyes_df['# of Frames Wrong'].mean()))/150)*100)
```

```python
beard_df
```

```python
beard_df[['True Gender']].value_counts(sort=False).plot.pie(labels=('Women','Men'),autopct='%1.1f%%')
plt.title('Gender')
plt.show()
```

```python
result=beard_df[((beard_df['True Gender']=='F') & (beard_df['Predict Gender']=='Woman')) | 
         ((beard_df['True Gender']=='M') & (beard_df['Predict Gender']=='Man'))]
result=len(result)
accuracy=result/50
print(accuracy*100,'%')
# Age... ~70% accuracy?
```

```python
wrong=len(shades_df[shades_df['Predict (%)']==0])
print('Accuracy rate of',((20-wrong)/20)*100,'%')
```
